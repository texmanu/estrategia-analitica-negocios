```python
cols_entrenar = ['deposit']
cols_dashboard = [cols for cols in data_train.columns if cols not in data_predict.columns]

for col in data_train.columns:
    # Agregar las columnas comunes entre data_train y data_predict
    if col in data_predict.columns and col not in cols_entrenar:
        cols_entrenar.append(col)


cols_entrenar, cols_dashboard
```

```python
# Generar el reporte de Sweetviz con la variable objetivo objetivo deposit
report = sv.analyze(data_train, target_feat='deposit')
# Generar el reporte de Sweetviz para las columnas del dashboard
report_dashboard = sv.analyze(data_train)
# Mostrar el reporte
report.show_html('report_entrenamiento.html')
```
